```python
def reset_table(df_,length):
    ddf = df_.iloc[length:,:]
    del ddf["00"]
    ddf.insert(4,"0:00",ddf["0:00:00"])
    del ddf["0:00:00"]
    ddf.rename(columns={"0:00":"0:00:00"},inplace=True)
    
    rdf = df_.iloc[:length-1,:]
    del rdf["0:00:00"]
    rdf.rename(columns={"00":"0:00:00"},inplace=True)
    df_1 = pd.concat([rdf,ddf],axis=0)
    return df_1
```

```python
def merge_df():
    merge = pd.DataFrame(pd.concat([df,df1,df2,df3,df5],axis=0))
    col_name = merge.columns.tolist()
    col_name[4] = "00"
    merge.columns = col_name
    df7 = reset_table(merge,len(df))
    return df7
```
