```python
def reset_table(df_):
    ddf = df_.iloc[14525:30017,:]
    del ddf["0:00:00"]
    ddf.insert(4,"0:00:00",ddf["00:00:00"])
    del ddf["00:00:00"]
#     ddf.rename(columns={"0:00:00":"0:00:00"},inplace=True)
    
    rdf_0 = df_.iloc[:14524,:]
    del rdf_0["00:00:00"]
#     rdf.rename(columns={"00:00:00":"0:00:00"},inplace=True)
     
    rdf_1 = df_.iloc[30018:,:]
    del rdf_1["00:00:00"]
#     rdf.rename(columns={"00":"0:00:00"},inplace=True)
    
    df_1 = pd.concat([rdf_0,ddf,rdf_1],axis=0,ignore_index=True)
    return df_1
```
