```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'hour_user_final.csv',encoding='utf8',index_col=0)
df
```

```python
df['model_ele'] = df.iloc[:,5:].apply(lambda x:x.sum(),axis=1)
```

```python
df['model_ele'] 
```

```python
idx = df[df['23:00:00']<0].index
# idx
df.drop(index=idx,inplace=True)
```

```python
df.insert(5,'model_ele1',df['model_ele'])
del df['model_ele']
df.rename(columns={"model_ele1":"model_ele"},inplace=True)
```

```python
df.iloc[:,6:] = df.iloc[:,6:].apply(lambda x:x/x.sum(),axis=1)
```

```python
df.drop(columns=['用电单元id','用电单元名称'],inplace=True)
```

```python
df.to_csv(r'user_model.csv',mode='w',encoding='utf8')
```

```python
df['时间'] = pd.to_datetime(df['时间'],format='%Y-%m-%d')
```

```python
df['时间'] = df['时间'].dt.month
df['时间']
```

```python
df.rename(columns={"时间":"月份"},inplace=True)
```
